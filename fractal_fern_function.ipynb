```python
# if r > 0.93: (x,y) -> (-0.15x + 0.28y, 0.26x + 0.24y + 0.44)
# if 0.86 < r <= 0.93: (x,y) -> (0.2x -0.26y, 0.23x + 0.22y + 1.6)
# if 0.01 < r <= 0.86: (x,y) -> (0.85x + 0.04y, -0.04x + 0.85y + 1.6)
# if r <= 0.01: (x,y) -> (0,0.16y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fractal_fern(n):
  n = n
  x = 0
  y = 0
  xn = 0
  yn = 0
  x1 = []
  y1 = []

  for i in range(n):
    r = np.random.default_rng()
    r = r.random()

    if r > 0.93:
      xn = (-0.15*x + 0.28*y)
      yn = (0.26*x + 0.24*y + 0.44)

    elif 0.86 < r <= 0.93:
      xn = (0.2*x -0.26*y)
      yn = (0.23*x + 0.22*y + 1.6)

    elif  0.01 < r <= 0.86:
      xn = (0.85*x + 0.04*y)
      yn = (-0.04*x + 0.85*y + 1.6)

    else:
      xn = (0)
      yn = (0.16*y)

    x = xn
    y = yn
    x1.append(x)
    y1.append(y)

  return x1, y1

```

```python
a = 35000

guardar = fractal_fern(a)
x0 = guardar[0]
y0 = guardar[1]

plt.scatter(x0, y0, color='g', s=1)
plt.show()
```
